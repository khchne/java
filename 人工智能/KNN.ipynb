```python
data_1 = data_1.values[:1000]
```

```python
n = len(data_1)//4
```

```python
test_set = data_1[:n]
train_set = data_1[n:]
```

```python
train_set
```

```python
def distance(d1, d2):
    return sum((d1 - d2)**2)**0.5
```

```python
def knn(data, k):
    res = [
        {"result":train[-1], "distance":distance(data[:-1], train[:-1])} for train in train_set
    ]
#     print(res)

    res = sorted(res, key=lambda items: items["distance"])
    res2 = res[:k]
#     print(res2)
    result = {True: 0, False: 0}
    sum = 0
    
    for r in res2:
        sum += r['distance']
    if sum == 0:
        return True
#     print(result[False])
#     print(sum)
    for r in res2:
#         print(r['result'])
#         print(result[r['result']])
        result[r['result']] += 1 - r['distance'] / sum
#         print("\n--> r = {} \n result[r['result']] = {}\n".format(r, result[r['result']]))
    
#     print(result)
#     print(data[-1])
#     return result
    if result[True]>result[False]:
        return True
    else:
        return False
    
```

```python
knn(test_set[0], 5)
```

```python
correctRate = 0
corrNum = 0
for i in test_set:
    res = knn(i, 5)
    if res == i[-1]:
        corrNum+=1
correctRate = 100*corrNum/len(test_set)
```

```python
correctRate
```

```python
corrRateList = []
x = list(range(1,20,2))
for k in range(1, 20, 2):
    correctRate = 0
    corrNum = 0
    for i in test_set:
        res = knn(i, k)
        if res == i[-1]:
            corrNum+=1
    correctRate = 100*corrNum/len(test_set)
    corrRateList.append(correctRate)
```

```python
plt.plot(x, corrRateList)
x = MultipleLocator(2)
# è®¾ç½®åˆ»åº¦é—´éš”
ax = plt.gca()
ax.xaxis.set_major_locator(x)
plt.xlim([0, 20])
plt.ylabel("Correct Rate: %")
plt.xlabel("The num of K")
plt.show()
```

```python
plt.plot(x, corrRateList)
plt.ylabel("Correct Rate: %")
plt.xlabel("The num of K")
plt.show()
```
